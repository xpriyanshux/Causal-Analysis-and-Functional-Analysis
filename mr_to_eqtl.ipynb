```python
import pandas as pd
import os

# === USER SETTINGS ===

# Path to your variant mapping file (with rsID and variant_id columns)
variant_map_path = "D:/final project/project work/eqtl mapping/variant_id_extraction/epilepsy.txt"

# Folder where GTEx *.signifpairs.txt files are located
gtex_eqtl_dir = "D:/diss2/gtex_eqtl/GTEx_Analysis_v8_eQTL/BRAIN/BRAIN"

# Folder to save filtered results
output_dir = "D:/final project/project work/eqtl mapping/epilepsy"
os.makedirs(output_dir, exist_ok=True)

# Significance threshold
pval_threshold = 1e-5

# === STEP 1: Load SNP mapping file ===
print("📥 Loading SNP mapping file...")
variant_map = pd.read_csv(variant_map_path, sep="\t")
rsid_to_variant = dict(zip(variant_map['rsID'], variant_map['variant_id']))
variant_to_rsid = {v: k for k, v in rsid_to_variant.items()}
variant_ids_set = set(variant_to_rsid.keys())
print(f"✅ Loaded {len(variant_ids_set)} variant IDs from mapping.")

# === STEP 2: Process each tissue file ===
print("\n🔄 Starting tissue-wise processing...\n")

for file_name in os.listdir(gtex_eqtl_dir):
    if not file_name.endswith(".signifpairs.txt"):
        continue

    tissue = file_name.replace(".signifpairs.txt", "")
    file_path = os.path.join(gtex_eqtl_dir, file_name)

    print(f"🔍 Processing: {tissue}")

    try:
        df = pd.read_csv(file_path, sep="\t")
    except Exception as e:
        print(f"❌ Failed to read {file_name}: {e}")
        continue

    if 'variant_id' not in df.columns or 'gene_id' not in df.columns:
        print(f"⚠️ Skipping {file_name}: required columns missing.")
        continue

    # Filter by matching variant IDs
    df = df[df['variant_id'].isin(variant_ids_set)]

    if df.empty:
        print(f"🚫 No matching SNPs found in {tissue}")
        continue

    # Add rsID column for interpretability
    df['rsID'] = df['variant_id'].map(variant_to_rsid)

    # Prioritize by p-value
    df_sig = df[df['pval_nominal'] < pval_threshold]

    if df_sig.empty:
        print(f"⚠️ No significant hits (p < {pval_threshold}) in {tissue}")
        continue

    # Save filtered output
    out_file = os.path.join(output_dir, f"{tissue}_prioritized_eqtl.csv")
    df_sig[['rsID', 'variant_id', 'gene_id', 'pval_nominal', 'slope']].to_csv(out_file, index=False)
    print(f"✅ Saved {df_sig.shape[0]} significant eQTLs to {out_file}")

print("\n🎉 DONE: All tissues processed and saved to:", output_dir)
```

```python
import pandas as pd
import os

# Directory containing per-tissue eQTL files
input_dir = "D:/final project/project work/eqtl mapping/epilepsy"
output_file = "D:/final project/project work/eqtl mapping/adem_eqtl.csv"

all_dfs = []

# Loop through all CSVs
for file_name in os.listdir(input_dir):
    if not file_name.endswith("_prioritized_eqtl.csv"):
        continue
    
    tissue = file_name.replace("_prioritized_eqtl.csv", "")
    df = pd.read_csv(os.path.join(input_dir, file_name))
    df["tissue"] = tissue  # Add tissue column
    all_dfs.append(df)

# Merge all
merged_df = pd.concat(all_dfs, ignore_index=True)
merged_df.to_csv(output_file, index=False)

print(f"✅ Merged {len(all_dfs)} files. Output saved to: {output_file}")
```
